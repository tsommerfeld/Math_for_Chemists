# Normal distribution

A random variable (RV) showing a normal distribution is called a normal RV.

The package [scientific Python](https://docs.scipy.org/doc/scipy-0.16.1/reference/stats.html) called `scipy` provides the module `stats`, which collects many statistics methods. We shall use it for the more advanced tasks.

```python
import scipy.stats
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#
# Uncomment the following lines for floating plot windows
# import matplotlib
# matplotlib.use('Qt5Agg')
# %matplotlib qt5
```

### Define a normal RV in Python

```python
rv = scipy.stats.norm
```

Now that we have `rv`, we can use its 'methods'. For example `rv.pdf(x)` is the PDF of a standard normal distribution, but you can specify $\mu$ and $\sigma$ using the notation `rv.pdf(x,loc=mu,scale=sigma)`.

The great thing about `scipy` is that I can define many other distributions in exactly the same way. For example, `rv=scipy.stats.uniform` defines a uniform RV or `rv=scipy.stats.t` defines a RV following Student's t-distribution. Of course, for every distribution the parameters will look a bit different, but apart from that, the 'methods' will be identical. 

```python
rv.pdf(0.5)
```

```python
rv.pdf(1, loc=0.5, scale=2)
```

```python
mu=0.5
sigma=2
xs=np.linspace(mu-4*sigma, mu+4*sigma, num=200)
ys=rv.pdf(xs, loc=mu, scale=sigma)
plt.cla()
plt.plot(xs,ys,'-')
plt.show()
```

The CDF works exactly the same: `rv.cdf(x)` or `rv.cdf(x,loc=mu,scale=sigma)`.

```python
rv.cdf(0.0)
```

```python
rv.cdf(1, loc=0.2, scale=0.5)
```

I hope you can see that this is vastly superior to using the Table in Shafer & Zhang's Figure 5.3.1. However, sometimes you only have the table, so I suggest you practice using the table and check with Python.

```python
mu=0.5
sigma=2
xs=np.linspace(mu-4*sigma, mu+4*sigma, num=200)
ys=rv.cdf(xs, loc=mu, scale=sigma)
plt.cla()
plt.plot(xs,ys,'-')
plt.show()
```
